```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading my dataset

train = pd.read_csv(r"C:\Users\HP\Videos\Machine Learning\train.csv")

test = pd.read_csv(r"C:\Users\HP\Videos\Machine Learning\test.csv")
```

```python
# I am going to use only numeric variables

numeric = ['int16','int64','int32','float64','float16','float32']

train = train.select_dtypes(include = numeric)

test = test.select_dtypes(include = numeric)
```

```python
x = train.iloc[:, :4]

x.head()
```

```python
# Importing Catboost

from catboost import CatBoostClassifier

X = train.drop('Promoted_or_Not', axis = 1)

y = train.Promoted_or_Not



from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
 X, y, test_size=0.3, random_state=42)

model = CatBoostClassifier(iterations = 1000, learning_rate = 1, depth = 2)
```

```python



model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

f1_score(pred, y_test)
```

```python
# Predicting on my test dataset

prediction = model.predict(test).astype(int)

prediction
```
