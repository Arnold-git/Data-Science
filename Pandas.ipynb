```python
#data frame in pandas

random_data = np.random.randn(8, 4)
print(random_data)
```

```python
#creating a data frame
#index = rows, column remains column

df = pd.DataFrame(random_data, index =['A', 'B','C','D','E','F','G','H'], columns = ['X','Y','Z','V'])
print(df)
```

```python
#selecting a column
#for two column pass it as a list

df[['X', 'Y']]
```

```python
# Creating a new column
#pass the name of column

df['N'] = df['X'] + df['V']
```

```python
df
```

```python
type(df['X'])
```

```python
#removing a column 
# use drop
#axis = 1

df.drop(['Z'], axis = 1)
```

```python
df
```

```python
# use inplace method to ensure the dataframe is changed

df.drop(['X'], axis = 1, inplace = True)

df
```

```python
# to manipulate row axis = 0
# remember column = 1

df.drop('E', axis = 0)
```

```python
#to select a row use the 'loc' method

df.loc['B']
```

```python
# to select multiple rows pass them as a list


df.loc[['B','D']]
```

```python
#selecting the use indexing use the method 'iloc'


df.iloc[2]
```

```python
#to select of row and column pass two lists
# the first is the correspondin row while
#  the other is the corresponding column

df.loc[['B','C'],['Z','V']]
```

```python
#using conditional selection

df>0
```

```python
#using condition selection to print out 
#the actual values

df[df>0]
```

```python
#using multiple conditions
#use '&' no 'and' and '|' for 'or'

df[(df['Z']>0 & (df['V']>0))]
```
